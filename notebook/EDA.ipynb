The dataset the goal is to predict price of given diamond (Regression Analysis).

There are 10 independent variables (including id):

id : unique identifier of each diamond
carat : carat (ct.) refers to the unique of weight measurement used exclusively to weight gemstones and diamonds.
cut : quality of diamonds cut
color : color of diamond
clarity : Diamond clarity is a measure of the purity and rarity of the stone , graded by the visibility of these characteristics 
under 10-power magnification
depth : The depth of diamond is its height(in millimeter) measured from the culet
(bottom tip) to the table (flat, top surface)
table : A diamond's table is the facet which can be seen when the stone is  viewed face up.
x: Diamond x diamension
y: Diamond y diamension
x: Diamond z dimension
